```python
import pyautogui
import time
import pytesseract
import pyperclip
import pandas as pd 

k = pd.read_csv('./Database/database.csv')


def problem_solved():
    r=k['Question']
    time.sleep(7)
    im = pyautogui.screenshot('./Image/test1.png',region=(37,201, 1388, 90))
    text = pytesseract.image_to_string("./Image/test1.png")
    text = text.split('.')[0]
    print(text)
    ans = r[int(text)-1]
    print(ans)
    pyautogui.click(x=1139, y=247, clicks=1, button='left')
    time.sleep(3)
    pyautogui.hotkey('command','a') 
    time.sleep(1)
    pyautogui.hotkey('command','x') 
    time.sleep(1)
    pyperclip.copy(ans)
    time.sleep(1)
    pyautogui.hotkey('command','v') 
    time.sleep(1)
    pyautogui.click(x=1379, y=440, clicks=1, button='left')
    time.sleep(7)
    pyautogui.click(x=234,  y=19, clicks=1, button='left')
for i in range(0,10):
    problem_solved()
    time.sleep(5)
# print(pyautogui.position())
```
